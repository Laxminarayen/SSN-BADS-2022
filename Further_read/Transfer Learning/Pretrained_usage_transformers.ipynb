```python
from sentence_transformers import SentenceTransformer
from transformers import pipeline
import spacy
import neuralcoref
from transformers import T5Tokenizer, T5ForConditionalGeneration, T5Config
import torch

model_sum = T5ForConditionalGeneration.from_pretrained('t5-small')
tokenizer = T5Tokenizer.from_pretrained('t5-small')
device = torch.device('cpu')

nlp = spacy.load('en_core_web_lg')
neuralcoref.add_to_pipe(nlp)
```

```python
import nltk
nltk.download('all')
```

```python
def preprocess(text):
  doc = nlp(text)
  resolved_text = doc._.coref_resolved
  return resolved_text
```

```python
resolved_text = preprocess(text)
```

```python
print(resolved_text)
```

```python
t5_prepared_text = 'summarize: '+resolved_text
tokenized_text = tokenizer.encode(t5_prepared_text, return_tensors = 'pt').to(device)
summary_ids = model_sum.generate(tokenized_text,max_length = 1000,min_length = 30)
output = tokenizer.decode(summary_ids[0],skip_special_tokens = True)
output
```
