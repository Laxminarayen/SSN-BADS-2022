```python
import re
text = "hi Inceptez is the best place to learn ML"
```

```python
text = "inceptez is the best place to learn ML"
z = re.match("Inceptez", text,re.IGNORECASE)
#This will just return the o/p as match object
print(z)

#To get the string 
print(z.group())
```

```python
z = re.search("Inceptez", text)
z
```

```python
z.group()
```

```python
#Without escaping case
text = "inceptez is the best place to learn ML. inceptez , Inceptez"
z = re.findall("Inceptez", text)
z
```

```python
text = "inceptez is the best place to learn ML. inceptez , Inceptez"
z = re.findall("Inceptez", text,re.I)
z
```

```python
import requests

data = requests.get("http://news.bbc.co.uk/2/hi/health/2284783.stm")
content = data.text
print(content)
```

```python
content
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(content,'html.parser')
```

```python
[s.extract for s in soup(['iframe','script'])]
```

```python
stripped_text = soup.get_text()
```

```python
print(stripped_text)
```

```python
stripped_text = re.sub(r'[\r|\n|\n\r]+',' ',stripped_text)
```

```python
stripped_text = re.sub(' +',' ',stripped_text)
```

```python
re.sub('[^A-Za-z0-9: ]+','',stripped_text)
```

```python
re.sub('\W+ ','',stripped_text)
```
